#### Menghapus value pada Dictionary

````python
profile2 = {
    'nama': ['dadan','didin','dodon','dudun','deden'] ,
    'umur': [30 , 40 , 26 , 25 , 33],
    'hobi': ['memasak','berkebun','basket','musik','kayang'] ,
    'sepatu': ['nike','adidas','puma','rebook','Asics'] }
````

````python
profile2['nama'].pop()
profile2
````

````python
profile2['nama'].remove('didin')
profile2
````

**Catatan**
1. del --> Berdasarkan index
2. remove --> berdasarkan value

````python
# Bisa menggunakan del
del profile2["hobi"][1]

# Atau pilih menggunakan .remove()
profile2["sepatu"].remove("puma")
````

````python
# Menyimpan file JSON
with open('data_NASA.json','w') as write_file: #w = write (export)
    json.dump(json_data,write_file)
````
