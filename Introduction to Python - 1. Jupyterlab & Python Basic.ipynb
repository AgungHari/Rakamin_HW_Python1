#### Tipe Data

| Kategori | Tipe | Keterangan | Contoh Kode Python |
| ----------- | ----------- | ----------- | ----------- |
| Text | str | Menyatakan karakter/kalimat bisa berupa huruf angka, dll (diapit dengan " atau ') | "Bahasa Python" |

| Boolean | bool | Menyatakan benar True atau salah False | True atau False |
| Sequence | list | Menyatakan beberapa untaian data dalam berbagai tipe data dan isinya bisa diubah-ubah (mutable) | ["Bakso", 21, True] |
| Sequence | tuple | Menyatakan beberapa untaian data dalam berbagai tipe data namun isinya tidak bisa diubah-ubah (immutable) | ("Bakso", 21, True) |
| Mapping | dict | Menyatakan beberapa untaian data dalam berbagai tipe data dan memiliki key/petujuntuk dalam menyimpan setiap nilai nya | {"nama":"Adam", "umur":24} |
| Set | set | Menyatakan beberapa untaian data dalam berbagai tipe data, setiap nilainya bernilai unik, dan tidak berurutan | {"sate", "bakso", "mie ayam"} |

<center>Tentang Penomoran/Indexing dalam string di Python</center>

| String | p | a | i | m | o | n |
| -- | -- | -- | -- | -- | -- | -- |
| Index | 0 | 1 | 2 | 3 | 4 | 5 |

<p></p>
<center>Untuk slicing rumusnya <b>variable</b><code>[index mulai:index akhir]</code>, cara baca <b>index akhir</b> ini selalu dikurangi oleh 1.
<p></p>    
Misal index mulai 2, index akhir 5, artinya kita mengambil urutan dari 2 hingga sebelum 5, yaitu 4, jadinya kita dapatkan pada index <b>2,3,4</b></center>

```python
# Slicing dalam String

nama = "paimon"
# String = p | a | i | m | o | n
# Index =  0 | 1 | 2 | 3 | 4 | 5

# Start index nya 2, end index nya 5:
# Artinya kita akan mengambil indeks pada 2,3, and 4 (sampai sebelum 5)
print(nama[2:5])

# Start Indexnya kosong, kemudian end index nya 3
# Artinya, kita ambil semua dari awal hingga selesai di indeks nomor 2 (sampai sebelum 3)
print(nama[ :3])

# Start index 2, kemudian end index nya kosong
# Artinya kita ambil mulai dari indeks 2, hingga seluruhnya (akhir)
print(nama[2: ])

# Kita balik si urutan huruf-hurufnya
print(nama[::-1])
```

#### Boolean

**Logical Operation (Boolean)**

| X | Y | NOT X | X OR Y | X AND Y |
| -- | -- | -- | -- | -- |
| True | True | False | True | True |
| True | False | False | True | False |
| False | True | True | True | False |
| False | False | True | False | False |

Terkait penggunaan <b>and</b> dan <b>or</b>:
1. <b>and</b>
   - Mengembalikan hasil **True** apabila <u>X dan Y adalah **True**</u>
   - Mengembalikan hasil **False** apabila <u>diantara X atau Y ada yang **False**</u>

2. <b>or</b>
   - Mengembalikan hasil **True** apabila <u>diantara X atau Y ada yang **True**</u>
   - Mengembalikan hasil **False** apabila <u>X dan Y adalah **False**</u>
